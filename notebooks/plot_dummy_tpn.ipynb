```python
from matplotlib import pyplot as plt
```

```python
x = [0, .1, .3,.4,.6,.7,.9,1]
y = [0, .2, .3,.2,.4,.3,.3,0]
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlim([0, 1])
ax.set_ylim([0,1])
ax.set(xlabel='k value (share of positive attribution scores)', ylabel='output score',
       title='Example of computation of TPN')
ax.grid()

fig.savefig("data/tpn_example_calculation.png")
plt.show()
```
