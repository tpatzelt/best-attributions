```python
x = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
y1 = [45, 53, 49, 85, 195, 85, 484, 221, 611, 945]
y2 = [1345, 1915, 1440, 1173, 1165, 1313, 1868, 1917, 2783, 3274]
y1_keyword = "feature attribution XAI"
y2_keyword = "XAI"
```

```python
fig, ax = plt.subplots()
ax.plot(x,y1, label=y1_keyword)
ax.plot(x,y2, label=y2_keyword)

ax.set(xlabel='year', ylabel='publications',
       title='Publications per Year (dimensions.ai)')
ax.grid()
ax.legend()

fig.savefig("data/xai_citations.png")
plt.show()
```
