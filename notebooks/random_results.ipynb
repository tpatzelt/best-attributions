```python
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open("../data/random_one_sample-softmax.json", "r") as fp:
    one_sample_softmax_results = json.load(fp)
    
with open("../data/random_all_samples-softmax.json", "r") as fp:
    all_sample_softmax_results = json.load(fp)
```

```python
plt.rcParams["figure.figsize"] = (20,3)
```

```python
random_tpns = one_sample_softmax_results['random_tpn_results']
random_tpss = one_sample_softmax_results['random_tps_results']
lig_tpn = one_sample_softmax_results['lig_tpn'][0]
lig_tps = one_sample_softmax_results['lig_tps'][0]

fig, ax = plt.subplots()

ax.plot(random_tpns)
ax.plot([lig_tpn] * len(random_tpns))
#ax.set_xlim([0, 1])
#ax.set_ylim([0,1])
ax.set(xlabel='iteration', ylabel='tpn score',
       title='TPN on Random Attribution Values for one sample')
#ax.grid()

#fig.savefig("data/tpn_example_calculation.png")
plt.show()

fig, ax = plt.subplots()

ax.plot(random_tpss)
ax.plot([lig_tps] * len(random_tpns))
#ax.set_xlim([0, 1])
#ax.set_ylim([0,1])
ax.set(xlabel='iteration', ylabel='tpn score',
       title='TPS on Random Attribution Values for one sample')
#ax.grid()

#fig.savefig("data/tpn_example_calculation.png")
plt.show()
```

```python
# count how many times the random attributions were better
random_tpss = np.array(random_tpss)
random_tpns = np.array(random_tpns)

perc_random_tps_better = np.mean(np.array(random_tpss) < lig_tps)
perc_random_tpn_better  = np.mean(np.array(random_tpns) < lig_tpn)
print(f"TPN of random attributions was in {perc_random_tpn_better*100}% of the samples better.")
print(f"TPS of random attributions was in {perc_random_tps_better*100}% of the samples better.")
```

```python
print(np.mean(random_tpns), np.mean(random_tpss))
mean_diff_tpn = np.mean(random_tpns - lig_tpn)
mean_diff_tps = np.mean(random_tpss - lig_tps)
print(f"Random attribution have on average a difference of TPN score of {mean_diff_tpn} and of TPS score of {mean_diff_tps}.")
```

```python
random_tpns, lig_tpns = zip(*all_sample_softmax_results['tpn_results'])
random_tpss, lig_tpss = zip(*all_sample_softmax_results['tps_results'])

fig, ax = plt.subplots()

ax.plot(random_tpns)
ax.plot(lig_tpss)
#ax.set_xlim([0, 1])
#ax.set_ylim([0,1])
ax.set(xlabel='iteration', ylabel='tpn score',
       title='TPN on Random Attribution Values and LIG')
#ax.grid()

#fig.savefig("data/tpn_example_calculation.png")
plt.show()

fig, ax = plt.subplots()

ax.plot(random_tpss)
ax.plot(lig_tpss)
#ax.set_xlim([0, 1])
#ax.set_ylim([0,1])
ax.set(xlabel='iteration', ylabel='tpn score',
       title='TPS on Random Attribution Values and LIG')
#ax.grid()

#fig.savefig("data/tpn_example_calculation.png")
plt.show()
```
